# Step 2: Define Output Schemas (Pydantic)

````python
from pydantic import BaseModel, Field
from typing import List, Dict, Optional

# Agent 1 Output Schema
class ProjectHistory(BaseModel):
    project: str
    role: str
    tech: List[str]

class DeveloperProfile(BaseModel):
    developer_id: str
    name: str
    primary_skills: List[str]
    secondary_skills: List[str]
    domains: List[str]
    experience_years: int
    project_history: List[ProjectHistory]
    confidence_score: Dict[str, float]

# New model to wrap a list of DeveloperProfile objects
class DeveloperProfilesList(BaseModel):
    profiles: List[DeveloperProfile]

# Agent 2 Output Schema
class ProjectTask(BaseModel):
    task_id: str
    title: str
    description: str
    required_skills: List[str]
    complexity: str
    dependency: List[str]
    task_type: str

class TaskList(BaseModel):
    tasks: List[ProjectTask]

# Agent 3 Output Schema
class DailyTask(BaseModel):
    task_id: str
    title: str
    allocated_hours: int
    work_description: str

class DaySchedule(BaseModel):
    day: str
    tasks: List[DailyTask]

class DeveloperSchedule(BaseModel):
    developer: str
    daily_capacity_hours: int
    schedule: List[DaySchedule]

class WeeklyAllocation(BaseModel):
    week: str
    developers: List[DeveloperSchedule]



#cv, project path
class RequestData(BaseModel):
    cv_dir: str
    proj_dir: str
    api_key: str



# Put them in a dictionary to pass to the factory
schemas = {
    'DeveloperProfilesList': DeveloperProfilesList,
    'TaskList': TaskList,
    'WeeklyAllocation': WeeklyAllocation
}
````

````python
def load_data_from_uploads(uploaded_files: List[UploadFile], source_label: str):
    """Processes a list of uploaded PDF files and returns LangChain Documents."""
    documents = []
    for file in uploaded_files:
        if file.filename.endswith(".pdf"):
            # Read the file content into memory
            content = file.file.read()
            pdf_reader = PdfReader(io.BytesIO(content))
            text = ""
            for page in pdf_reader.pages:
                text += page.extract_text()

            documents.append(Document(
                page_content=text,
                metadata={"source": source_label, "file": file.filename}
            ))
    return documents
````

# Splitter  and Embedding

````python
from langchain_text_splitters import RecursiveCharacterTextSplitter
from langchain_huggingface import HuggingFaceEmbeddings
from langchain_community.vectorstores import Chroma
from langchain_core.documents import Document


def create_knowledge_base(documents):
    """Splits documents and initializes the Vector Database."""
    text_splitter = RecursiveCharacterTextSplitter(
        chunk_size=2000,
        chunk_overlap=200,
        separators=["\n\n", "\n", ".", " "]
    )
    chunks = text_splitter.split_documents(documents)
    return chunks
````

# Embedding and Vector Database

````python
from langchain_google_genai import GoogleGenerativeAIEmbeddings
from langchain_community.vectorstores import Chroma
import uuid
import traceback

def create_vector_store(chunks, persist_dir, api_key):
    """
    Stage 2: Logic for embedding and storing chunks.
    This is where the Google API is called.
    """
    print(f"🧠 Generating embeddings for {len(chunks)} chunks...") # Add this to see the count!
    embeddings = GoogleGenerativeAIEmbeddings(
        model="models/gemini-embedding-001",
        google_api_key=api_key,
        # Reduce this if you are on a free tier to avoid 429 Errors
        request_options={'timeout': 5000}
    )

    vector_db = Chroma.from_documents(
        documents=chunks,
        embedding=embeddings,
        persist_directory=persist_dir
    )
    return vector_db

# Usage:
#db_cv = create_vector_store(cv_chunks, "./cv_chroma_db", gimini_key)
#db_project = create_vector_store(project_chunks, "./project_chroma_db", gimini_key)
````

# Tool

````python
from crewai.tools import BaseTool                            #factory approach
from pydantic import BaseModel, Field
from typing import Type

# 1. Define the Schema Class (The Object)
class SearchInput(BaseModel):
    query: str = Field(..., description="The search string to look for (e.g., 'Python expertise' or 'API requirements')")

def create_custom_tools(cv_db, project_db):
    """
    Factory function that injects specific VectorDB instances into tools.
    """

    class CVSearchTool(BaseTool):
        name: str = "search_cv_database"
        description: str = "Useful for finding developer skills, experience, and profiles from uploaded CVs."
        # We pass the class object here
        args_schema: Type[BaseModel] = SearchInput

        def _run(self, query: str) -> str:
            # Similarity search against the dynamically created CV database
            results = cv_db.similarity_search(query, k=3)
            return "\n".join([d.page_content for d in results])

    class ProjectSearchTool(BaseTool):
        name: str = "search_project_database"
        description: str = "Useful for extracting project requirements, tasks, and technical specs."
        # We pass the class object here
        args_schema: Type[BaseModel] = SearchInput

        def _run(self, query: str) -> str:
            # Similarity search against the dynamically created Project database
            results = project_db.similarity_search(query, k=3)
            return "\n".join([d.page_content for d in results])

    # Return initialized instances of the local classes
    return [CVSearchTool(), ProjectSearchTool()]
````

# Agent Building

````python
def get_optimized_lessons(LEARNING_FILE="agent_memory.json"):
    if not os.path.exists(LEARNING_FILE):
        return "No feedback yet."
    
    with open(LEARNING_FILE, "r") as f:
        history = json.load(f)
        
    # If the history is small, send it all
    if len(history) < 5:
        return "\n".join([f"- {h['user_insight']}" for h in history])
    
    # If history is large, we focus on the MOST IMPORTANT/RECENT insights
    # This prevents the LLM from getting "Information Overload"
    recent_lessons = history[-10:] 
    
    context = "Based on recent user feedback, here are your improved behaviors:\n"
    for h in recent_lessons:
        context += f"• When user said '{h['user_insight']}', they rated complexity as {h['complexity']}.\n"
        
    return context
````

````python
from crewai import Agent
def create_agents(tools, llm,session_memory=None):
    """
    Function-oriented Agent factory.
    Injects the specific tools and LLM for this specific API request.
    """
    # Unpack the tools from the factory we created in the previous step
    search_cv_database, search_project_database = tools
    
    # Example: "John Doe: 20 hours, Jane Smith: 35 hours"
    if session_memory:
        memory_context = ", ".join([f"{name}: {hrs} hours" for name, hrs in session_memory.items()])
    else:
        memory_context = "No previous tasks assigned in this session."
    
    # 2. Get the Past Lessons (Learning from feedback)
    past_lessons =get_optimized_lessons() # Call the helper function here
    

        
        
    # Agent 1: Lead Technical Recruiter
    lead_recruiter = Agent(
        role='Lead Technical Recruiter',
        goal='Analyze all CVs in the database and extract developer profiles.',
        backstory="""You are an expert recruiter. Your SOLE purpose is to extract developer profiles in a strict JSON object format, where the list of profiles is under a top-level key named 'profiles'.
        When using the search_cv_database tool, provide ONLY the query string (Format: {\"query\": \"search terms\"}).
        Your final answer MUST be ONLY a JSON object conforming to the DeveloperProfilesList schema, with ABSOLUTELY NO other text, conversational elements, explanations, or MARKDOWN CODE BLOCKS (```json). Just the JSON object.
        Do not use nested properties or types in tool inputs.""",
        tools=[search_cv_database],
        llm=llm,
        verbose=True,
        allow_delegation=False,
        max_iter=5
    )

    # Agent 2: The Project Architect
    project_analyzer_agent = Agent(
        role='Senior Systems Architect',
        goal='Decompose project documentation into granular, actionable tasks.',
        backstory="""You are an expert at project estimation. Your SOLE purpose is to output a strict JSON object format, where the list of tasks is under a top-level key named 'tasks'.
        When using the search_project_database tool, provide ONLY the query string (Format: {\"query\": \"project requirements\"}).
        Your final answer MUST be ONLY a JSON object conforming to the TaskList schema, with ABSOLUTELY NO other text, conversational elements, explanations, or MARKDOWN CODE BLOCKS (```json). Just the JSON object.
        Do not use nested properties.""",
        tools=[search_project_database],
        llm=llm,
        verbose=True,
        allow_delegation=False,
        max_iter=5
    )


    

    # Agent 3: The Task Allocator (CRITICAL CHANGES HERE)
    allocation_agent = Agent(
        role='Engineering Manager',
        goal='Match developers to tasks based on skill-fit and remaining session capacity and also incorporate past lessons learned.',
        backstory=f"""
        You oversee the engineering team. You are a 'Learning Agent
        You take the developer profiles and the task list to create the final assignments.
        Your final answer MUST be ONLY a JSON object conforming to the WeeklyAllocation schema, 
        with ABSOLUTELY NO other text, conversational elements, explanations, or 
        MARKDOWN CODE BLOCKS (```json). Just the JSON object.
        
        
        USER FEEDBACK HISTORY:
        {past_lessons}
        
        CURRENT OCCUPIED HOURS see details in memory_context (DO NOT EXCEED 35h TOTAL):
        CURRENT SESSION MEMORY: {memory_context}
    
        CRITICAL INSTRUCTIONS:
        1. Total hours (Existing + New) per person MUST NOT exceed 35.
        2. If a developer in 'CURRENT OCCUPIED HOURS' is at 35, they are FULL.
        3. If all developers are full, return an empty 'assignments' list and 
           state 'All developers are currently occupied' in the result summary.
        4. If 'SESSION MEMORY' shows a dev is at 35h, they are UNAVAILABLE.
        Your final answer MUST be ONLY a JSON object conforming to the WeeklyAllocation schema.""",
        llm=llm,
        verbose=True,
        allow_delegation=False,
        max_iter=5
    )
    return lead_recruiter, project_analyzer_agent, allocation_agent
````

# Task

````python
from crewai import Task

def create_tasks(lead_recruiter, project_analyzer_agent, allocation_agent, schemas,callback):
    """
    Function-oriented Task factory.
    Links tasks to the specific agent instances and Pydantic schemas.
    """
    # Unpack schemas for easy access
    DeveloperProfilesList = schemas['DeveloperProfilesList']
    TaskList = schemas['TaskList']
    WeeklyAllocation = schemas['WeeklyAllocation']

    # Task 1: CV Extraction
    task1 = Task(
        description="""Search the CV database. For each developer found, extract:
        - name, primary_skills, secondary_skills, and experience_years.
        - Limit 'project_history' to the 2 most relevant projects.
        Return a SINGLE JSON object with the key 'profiles'.
        DO NOT include any text before or after the JSON.""",
        expected_output="A complete, valid JSON object following the DeveloperProfilesList schema.",
        agent=lead_recruiter,
        output_pydantic=DeveloperProfilesList
    )

    # Task 2: Project Decomposition
    task2 = Task(
        description="Analyze the project document and generate a list of weekly tasks.",
        expected_output="A structured list of technical tasks.",
        agent=project_analyzer_agent,
        output_pydantic=TaskList
    )

    # Task 3: Weekly Ticket Generation
    task3 = Task(
        description="""
        1. Review the developer profiles from Task 1 (context[0]).
        2. Review the project tasks from Task 2 (context[1]).
        3. For each task, find a developer whose 'primary_skills' match the 'required_skills'.
        4. Distribute tasks across a 5-day week (Mon-Fri).
        5. Ensure no developer exceeds 7 hours of work per day.
        6. Complexity of Task
        7. Return the final allocation as a WeeklyAllocation JSON object.
        """,
        expected_output="A complete WeeklyAllocation Pydantic object containing the full schedule.",
        agent=allocation_agent,
        context=[task1, task2],
        output_pydantic=WeeklyAllocation
    )

    return [task1, task2, task3]
````

````python
from crewai import Crew, Process

def run_sprint_crew(agents, tasks):
    """
    Function-oriented Crew execution.
    Combines agents and tasks into a process and returns the final Pydantic result.
    """

    # Initialize the Crew
    weekly_crew = Crew(
        agents=agents,
        tasks=tasks,
        process=Process.sequential,
        verbose=True
    )

    # Execution starts here
    # Use kickoff() for standard scripts or akickoff() for FastAPI
    result = weekly_crew.kickoff()
    
    
    # Save the final structured output to a JSON file
    with open("weekly_sprint_plan.json", "w") as f:
        f.write(result.pydantic.model_dump_json(indent=2))
        print("Weekly tickets generated successfully!")

    return result
````

````python
import io
from PyPDF2 import PdfReader
from langchain_core.documents import Document

def process_bytes_to_docs(data_list, label):
    """
    Converts a list of dicts [{'filename': str, 'content': bytes}] 
    into a list of LangChain Document objects.
    """
    documents = []
    for item in data_list:
        try:
            # Wrap bytes in BytesIO so PdfReader treats it like a file
            file_stream = io.BytesIO(item['content'])
            reader = PdfReader(file_stream)
            
            # Extract text from every page
            text = ""
            for page in reader.pages:
                extracted_text = page.extract_text()
                if extracted_text:
                    text += extracted_text + "\n"
            
            # Create the Document object for the Vector DB
            documents.append(Document(
                page_content=text,
                metadata={
                    "source": label,
                    "filename": item['filename']
                }
            ))
        except Exception as e:
            print(f"Error processing {item['filename']}: {e}")
            continue # Skip corrupted files
            
    return documents
````

# Fast API

````python
#main.py
from fastapi import FastAPI, HTTPException
from fastapi import BackgroundTasks
import uuid
import traceback
from typing import List, Dict, Optional
from fastapi import UploadFile, File
import json
from datetime import datetime



app = FastAPI(title="Weekly Sprint Planner API")
# Hardcode your keys here (Keep them safe!)
MY_HF_TOKEN = "your key"
MY_GEMINI_KEY = "your key"
LEARNING_FILE = "agent_memory.json"


# Global storage for results (In production, use Redis or a Database)
jobs_db = {}
session_allocations = {}  #for maintianing  a record of which developer is allocated to which task in which session. This can be used for tracking and future reference.

@app.get("/")
def read_root():
    return {"Hello": "World"}



def update_session_hours(result_data):
    """
    Parses the 'WeeklyAllocation' result and updates global memory.
    """
    try:
        # Check both potential keys: 'assignments' or 'profiles'
        assignments = result_data.get('assignments') or result_data.get('profiles') or []
        
        for dev in assignments:
            # Get the developer name
            name = dev.get('developer_name') or dev.get('name')
            if not name: continue
            
            # Sum up hours from the schedule list
            # We use int() to ensure math works, defaulting to 0 if missing
            schedule = dev.get('schedule') or dev.get('tasks') or []
            new_job_hours = sum(int(item.get('hours', 0)) for item in schedule)
            
            # Add to global memory
            if name in session_allocations:
                session_allocations[name] += new_job_hours
            else:
                session_allocations[name] = new_job_hours
                
            print(f"📊 Session Update: {name} now has {session_allocations[name]} total hours.")
            
    except Exception as e:
        print(f"⚠️ Warning: Could not parse hours for session update: {e}")
        
        

# --- UPDATED PIPELINE ---
async def run_full_ai_pipeline(job_id: str, cv_data, proj_data, schemas):
    jobs_db[job_id] = {
        "status": "Processing",
        "progress_logs": ["🚀 Initializing pipeline..."],
        "result": None,
        "error": None
    }
    
    def update_log(message):
        if job_id in jobs_db:
            jobs_db[job_id]["progress_logs"].append(message)
            print(f"LOG [{job_id}]: {message}")

    try:
        update_log("📄 Extracting text and Creating searchable chunks...")
        raw_cv_docs = process_bytes_to_docs(cv_data, "cv")
        raw_proj_docs = process_bytes_to_docs(proj_data, "project")
        cv_chunks = create_knowledge_base(raw_cv_docs)
        proj_chunks = create_knowledge_base(raw_proj_docs)
        
        update_log("🧠 Generating Gemini Embeddings...")
        db_cv = create_vector_store(cv_chunks, f"./db/cv_{job_id}", MY_GEMINI_KEY)
        db_proj = create_vector_store(proj_chunks, f"./db/proj_{job_id}", MY_GEMINI_KEY)

        update_log("🛠️ Setting up Agents with Session Memory...")
        llm_instance = create_llm(MY_HF_TOKEN)
        tools = create_custom_tools(db_cv, db_proj)
        
        # --- KEY CHANGE: PASS SESSION MEMORY TO AGENTS ---
        agents = create_agents(tools, llm_instance, session_memory=session_allocations)

        def task_completed_callback(task_output):
            update_log(f"✅ Task completed by Agent.")

        tasks = create_tasks(
            lead_recruiter=agents[0], 
            project_analyzer_agent=agents[1], 
            allocation_agent=agents[2], 
            schemas=schemas,
            callback=task_completed_callback
        )

    
        # --- STAGE 6: FINALIZATION ---
        update_log("🏁 Success! Saving data to session memory...")
        result = run_sprint_crew(agents, tasks)
        
        # 1. Convert Pydantic to Dict
        final_dict = result.pydantic.model_dump()
        
        # 2. Update the Global Database
        update_session_hours(final_dict)

        # 3. Update the job status for the Frontend
        jobs_db[job_id].update({
            "status": "Completed",
            "result": final_dict,
            "session_status": session_allocations # Pass current totals to UI
        })
    except Exception as e:
        print(traceback.format_exc()) 
        update_log(f"❌ Error: {str(e)}")
        jobs_db[job_id].update({"status": "Failed", "error": str(e)})






@app.post("/generate-plan")
async def handle_request(
    background_tasks: BackgroundTasks,
    cv_files: List[UploadFile] = File(...),
    project_files: List[UploadFile] = File(...)
):
    # 1. Create unique ID
    job_id = str(uuid.uuid4())
    
    # 2. Initialize the DB immediately
    # This prevents 'KeyError' when the background task starts logging
    jobs_db[job_id] = {
        "status": "Processing", 
        "progress_logs": ["📥 Files received by server..."], 
        "result": None, 
        "error": None
    }

    try:
        # 3. CRITICAL: Read the files into memory BEFORE the request finishes
        # This is the "Bytes" part that solves the 'I/O on closed file' error
        cv_bytes_data = [
            {"filename": f.filename, "content": await f.read()} 
            for f in cv_files
        ]
        proj_bytes_data = [
            {"filename": f.filename, "content": await f.read()} 
            for f in project_files
        ]
    except Exception as e:
        jobs_db[job_id].update({"status": "Failed", "error": f"Upload Error: {str(e)}"})
        raise HTTPException(status_code=400, detail=f"File read error: {e}")

    # 4. Offload to background
    # Note: We pass the BYTES data, not the original 'cv_files' objects
    background_tasks.add_task(
        run_full_ai_pipeline, 
        job_id, 
        cv_bytes_data, 
        proj_bytes_data,
        schemas
    )

    return {
        "job_id": job_id,
        "status": "Accepted",
        "message": "AI Agents have started working. Check status via job_id.",
        "check_status_url": f"/status/{job_id}"
    }



# --- UPDATED STATUS ENDPOINT ---
@app.get("/status/{job_id}")
async def get_status(job_id: str):
    job = jobs_db.get(job_id)
    if not job:
        raise HTTPException(status_code=404, detail="Job ID not found")
    
    # Inject the global session data so the Frontend can see the progress bars
    job["session_status"] = session_allocations
    return job



@app.post("/learn")
async def save_feedback(data: dict):
    try:
        memory = []
        if os.path.exists(LEARNING_FILE):
            with open(LEARNING_FILE, "r") as f:
                memory = json.load(f)
        
        # We capture the RAW note so the Agent can interpret the "intent"
        memory.append({
            "timestamp": datetime.now().strftime("%Y-%m-%d %H:%M:%S"),
            "user_insight": data.get("user_insight"),  # What the user actually said
            "timing": data.get("timing"),
            "complexity": data.get("complexity")
        })
        
        with open(LEARNING_FILE, "w") as f:
            json.dump(memory, f, indent=4)
        return {"status": "success"}
    except Exception as e:
        raise HTTPException(status_code=500, detail=str(e))


    
    
    
    

@app.post("/reset-session")
async def reset_session():
    """
    Clears the global memory of assigned hours.
    """
    global session_allocations
    session_allocations = {}
    return {"status": "success"}

````
